# Testing Mikeys Segementation Method

This method provides channels with very small pixel values, I am going to go from pre-masked inputs to each segmented cell.

```python
import pandas as pd 
import numpy as np
from typing import List

from PIL import Image
import matplotlib.pyplot as plt

# HPA
import os
import imageio
from hpacellseg.cellsegmentator import CellSegmentator
from hpacellseg.utils import label_cell
```

```python
df = pd.read_csv('../input/image_subset_example/example.csv')
df
```

```python
test_image = df.loc[1, :]
test_ID = test_image.ID
```

```python
def read_image_array(ID, path='../input/image_subset_example/image/'):
    r = plt.imread(path+ID+'_red.png')
    g = plt.imread(path+ID+'_green.png')
    b = plt.imread(path+ID+'_blue.png')
    y = plt.imread(path+ID+'_yellow.png')
    return r, g, b, y

r, g, b, y = read_image_array(test_ID)

# ref_channels are r y b

ref_channels = [r, y, b]
```

```python
# model weights
NUC_MODEL = "../models/nuclei-model.pth"
CELL_MODEL = "../models/cell-model.pth"

seg = CellSegmentator(
    NUC_MODEL,
    CELL_MODEL,
    scale_factor=0.25,
    device="cpu",
    padding=False,
    multi_channel_model=True,
)
```

```python
def segment(ref_channels: List[str], segmentator: CellSegmentator):
    # Ref channels must be in order red, yellow, blue
    input_ = [[i] for i in ref_channels]  # Segmentator only accepts list of lists input
    nuc_segmentation = segmentator.pred_nuclei(input_[2])[0]
    cell_segmentation = segmentator.pred_cells(input_)[0]
    mask = label_cell(nuc_segmentation, cell_segmentation)[1]
    return mask
```

```python
# get mask
mask = segment(ref_channels, seg)
```

```python
image = np.dstack((r, g, b))
```

```python
print('Number of cells in mask (0 is background): {}'.format(np.unique(mask)))
fig, ax = plt.subplots(figsize=(10,15))
ax.imshow(image)
ax.imshow(mask, alpha=0.5)
ax.axis('off')
plt.show()
```

This is our target cell. Next we will only take where the temp mask == 1.

```python
# Get temp mask bounding box coords
# nonzero() returns indices of where elements are non-zero
idxs = np.asarray(temp_mask == 1).nonzero()
y_min, y_max = idxs[0].min(), idxs[0].max()
x_min, x_max = idxs[1].min(), idxs[1].max()
print('Indices where non zero shape {}'.format(np.shape(idxs)))
```

```python
cell_channels = []
```

So now we have the bounding box of the cell, we can then only return the cell channels of those bounding boxes.

We should expect cell_channels to have a length of 4 and be the size of the bounding boxes, then we can check the pixel values.

```python
# channel arrays in a list
channels = [r, g, b, y]

for channel_arr in channels:
    # Zero pad and square off
    single_cell = temp_mask * channel_arr
    single_cell = single_cell[y_min:(y_max + 1), x_min:(x_max + 1)]
    
    cell_channels.append(single_cell)
```
